## Using matplotlib
### Example sine function
$$z=sin(x^2+y^2)$$
One can extend this by adding a shift to the sine function:
$$z=sin(x^2+y^2 - a)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib inline½
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
fig, ax = plt.subplots()
images = []
for a in np.linspace(-np.pi, np.pi, 50):
    Z = np.sin(X ** 2 + Y ** 2 - a)
    img = ax.imshow(Z)
    images.append([img])
output = animation.ArtistAnimation(fig, images, interval=50, blit=True)
plt.show()
```
